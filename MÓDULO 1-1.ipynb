```
variavel3 = 'olá'
print(variavel3)
type(variavel3)
```

## Reatribuindo valores para uma lista já criada

```
x = ['zero', 'castle', 'forza']
x[1] = 'castlevania'
print(x)
type(x)
```

# Valor Tuple

```
x = ['john', 'cipriani', 'wick']
x[0] > 'john'
x[2] > 'wick'
x[0] = 'johnsson'
print(x)
type(x)
```
