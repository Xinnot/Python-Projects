```
def reverse_words(s):
    words = s.split(' ')
    words = reversed(words)
    words = " ".join(words)
    print(words)
    return words
```

```
def solution(a, b):
    if len(a) > len(b):
        ordem = b + a + b
        return ordem
    elif len(a) < len(b):
        ordem = a + b + a
        return ordem
```

```
def find_smallest_int(arr):
    menor = arr[0]
    for numero in arr:
        if numero < menor:
            menor = numero

    return menor
```
