```
def find_smallest_int(arr):
    return min(arr)
```

```
def find_average(nums):
    soma = 0
    if len(nums) == 0:
        return 0
    elif not nums:
        return 0
    else:
        for i in nums:
            soma = soma + i
            media = soma / len(nums)
        return media
```
