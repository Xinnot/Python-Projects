# Condições

Se o 3 é maior que 2, imprime 
* é maior
* foi

```
if 3 > 2:
  print('é maior')
  print('foi')
```

E se a nossa condição não for aceita, o que fazemos?

Utilizamos o 'else'

```
if 2 > 3:
  print('é maior')
  print('foi')
else:
  print('não é maior')
  print('foi também')
```
