```
def people_with_age_drink(age):
    if age < 14:
        return 'drink toddy'
    elif age < 18:
        return 'drink coke'
    elif age < 21:
        return 'drink beer'
    else:
        return 'drink whisky'
print(people_with_age_drink(19))
```

```
def hero(bullets, dragons):
    if bullets >= 2*dragons:
        return True
    else:
        return False
print(hero(3,2))
```
